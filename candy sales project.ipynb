```python
df['Order ID'] = df['Order ID'].astype('category')
df['Order Date'] = df['Order Date'].astype('category')
df['Ship Date'] = df['Ship Date'].astype('category')
df['Ship Mode'] = df['Ship Mode'].astype('category')
df['Country/Region'] = df['Customer ID'].astype('category')
df['City'] = df['City'].astype('category')
df['Product Name'] = df['Product Name'].astype('category')
```

```python
## check for the total rows and columns 
print(f'total number of rows: {df.shape[0]} => total number of columns: {df.shape[1]}')
```

```python
for each_name in df.columns: 
    print(each_name)
    print(df[each_name].unique())
```

```python
df['Row ID'] = df ['Row ID'].replace('...',np.nan)
df['Order ID'] = df ['Order ID'].replace('...',np.nan)
df['Order Date'] = df ['Order Date'].replace('...',np.nan)
df['Ship Date'] = df ['Ship Date'].replace('...',np.nan)
df['Customer ID'] = df ['Customer ID'].replace('...',np.nan)

```

```python
for each_name in df.columns: 
    print(each_name)
    print(df[each_name].unique())
```
